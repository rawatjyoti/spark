# To perform Exploratory data analysis

```python

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('SampleSuperstore.csv')

data.head()
```

```python
data.shape
```

# Data Preprocessing

```python
data.isnull().sum().sum()
data.dtypes
```

No need of preprocessing is required as there are no null values and all data types are correct

# Data Visualization

```python
sns.pairplot(data)
plt.show()
```

```python
plt.figure(1)
plt.subplot(221)
data['Category'].value_counts().plot.bar(figsize=(20,10), color='orange', title='Category')
plt.subplot(222)
data['Segment'].value_counts().plot.bar(color='violet', title='Segment')
plt.subplot(223)
data['Ship Mode'].value_counts().plot.bar(color='green', title='Shop Mode')
plt.subplot(224)
data['Region'].value_counts().plot.bar(color='yellow', title='Region')

plt.show()
```

It can be inferred from the above bar plots that:

Most of the Category are Supplies.
Most of the Segment are Consumer.
Most of the Ship Mode are of Standard Class.
Region is distributed but maximun is from West region.

```python
data.plot(kind='scatter', x='Profit', y='Sales', figsize=(14, 8), color='darkblue')
sns.regplot('Sales','Profit', data=data, color='red')
plt.show()
```

From the above Scatter and Regplot plot:

Profit is not solely depending upon Sales. There are many other factors acting upon it.

```python

data.plot(kind='scatter', x='Quantity', y='Profit', figsize=(12, 6), color='orange')
plt.show()
```

```python
sns.catplot(x="Sub-Category", y="Profit", data=data, color='Green', height=4, aspect=4);
```

From the above Catplot:

It can be viewed that Binders are higly varable and Copiers earns maximum Profit.

```python
data.plot.scatter('Discount', 'Profit', figsize=(12, 6), color='aqua')
plt.show()
```
