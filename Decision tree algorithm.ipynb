# Data Modelling

```python
y = iris['Species']
X = iris.drop('Species', axis=1)
```

```python
dt = DecisionTreeClassifier()
model = dt.fit(X,y)
```

# Visualizing Decision Trees

We can visualize Decision Trees using only a single Python library tree of sklearn.

```python
text = tree.export_text(dt)
```

```python
print(text)
```

```python
with open('decision_tree.log', 'w')as log:
    log.write(text)
```

```python
fig = plt.figure(figsize=(20,10))
tree.plot_tree(dt, feature_names=iris.columns, class_names=iris['Species'] ,filled=True)
```
